## Putting some Weights and testing the similarities between them

```python
# Create TFidfVectorizer 
tfidf_G= TfidfVectorizer()
tfidf_S= TfidfVectorizer()
tfidf_T= TfidfVectorizer()

# Fit and transform the document
tfidf_vector_G = tfidf_G.fit_transform(danganronpa_genres)
tfidf_vector2_G = tfidf_G.transform([danganronpa_genres[2]])
tfidf_vector_S = tfidf_S.fit_transform(danganronpa_synopsis)
tfidf_vector2_S = tfidf_S.transform([danganronpa_synopsis[2]])
tfidf_vector_T = tfidf_T.fit_transform(danganronpa_title)
tfidf_vector2_T = tfidf_T.transform([danganronpa_title[2]])
# Compute cosine similarity
cosine_sim_G=cosine_similarity(tfidf_vector_G, tfidf_vector2_G)
cosine_sim_S=cosine_similarity(tfidf_vector_S, tfidf_vector2_S)
cosine_sim_T=cosine_similarity(tfidf_vector_T, tfidf_vector2_T)
```

<font size=4>Here, as my personal taste, I always consider way more the <b>genres</b> first then, I give the <b>synopsis</b> a chance. <br>
So I have put some weight on them.<br><br>
    <font size=4>I can <b>NOT</b> totally ignore the <b>synopsis</b> or <b>title</b>, because them sometimes can give us a good sign of a new season for that anime or even some similarities in theme between them

```python
WT_G=4
WT_S=3
WT_T=2
cosine_sim_sum=(cosine_sim_G*WT_G+cosine_sim_S*WT_S+cosine_sim_T*WT_T)
cosine_sim_mean=cosine_sim_sum/(WT_G+WT_S+WT_T)
```

```python
cosine_sim_mean=cosine_sim_sum/9
```

<font size=4> Now let's take a look at where the cosine_sim_mean is above .5 and below 1 

```python
cosine_sim_mean[np.where((cosine_sim_mean<1) & (cosine_sim_mean>0.5))]
```

```python
you_watched=[]
for i in np.where((cosine_sim_mean<1) & (cosine_sim_mean>0.5))[0].tolist():
    print(danganronpa_title[i])
    you_watched.append(danganronpa_title[i])
```

<font size=4> I just sum all value in the cosine_sim_mean which were above 0.5 and below 1 and divided by its own length.<br> so there I could get the <b>mean of similarity for all titles</b> that were similar

```python
chance_sim_all_watched=np.sum(cosine_sim_mean[np.where((cosine_sim_mean<1) & (cosine_sim_mean>0.5))])/len(cosine_sim_mean[np.where((cosine_sim_mean<1) & (cosine_sim_mean>0.5))])
print(chance_sim_all_watched)
```

```python
print(f'You might like "{danganronpa_title[2]}"\n{chance_sim_all_watched*100:.2f}% chance of liking it due to others titles you watched:\n{you_watched[:5]}')
```

# Testing Similarities in the Seasonal Animes

```python
year=2022
season='winter'
```

```python
animes_season = pd.read_csv(f'./{year}_{season}_Preprocessed.csv')
animes_season.drop("Unnamed: 0",axis=1,inplace=True)
#animes.drop("",axis=1,inplace=True)
#animes.dropna(inplace=True)
pd.set_option('display.max_rows', animes_season.shape[0]+1)
print(animes_season.title)
```

<br><font size =4>Here, I Just create 2 lists based on chance_sim_all_watched:<br>
    <b>DEFINITELY WATCH:</b> chance_sim_all_watched > 60%<br>
    <b>MIGHT WATCH::</b> chance_sim_all_watched < 60%

```python
# Create TFidfVectorizer 
tfidf_G= TfidfVectorizer()
tfidf_S= TfidfVectorizer()
tfidf_T= TfidfVectorizer()

 # Fit and transform my animes watched
tfidf_vector_G = tfidf_G.fit_transform(docs_genres)
tfidf_vector_S = tfidf_S.fit_transform(docs_synop)
tfidf_vector_T = tfidf_T.fit_transform(docs_title)
c=0
DEFINITELY=[]
MIGHT=[]
for i in range(0,len(animes_season.title)):
   
    #print(animes_season.title[i])
    #Transform anime the new dataset of animes 
    tfidf_vector2_G = tfidf_G.transform([animes_season.genres[i]])
    tfidf_vector2_S = tfidf_S.transform([animes_season.synopsis[i]])
    tfidf_vector2_T = tfidf_T.transform([animes_season.title[i]])
    
    # Compute cosine similarity
    cosine_sim_G=cosine_similarity(tfidf_vector_G, tfidf_vector2_G)
    cosine_sim_S=cosine_similarity(tfidf_vector_S, tfidf_vector2_S)
    cosine_sim_T=cosine_similarity(tfidf_vector_T, tfidf_vector2_T)
    
    WT_G=4
    WT_S=3
    WT_T=2
    cosine_sim_sum=(cosine_sim_G*WT_G+cosine_sim_S*WT_S+cosine_sim_T*WT_T)
    cosine_sim_mean=cosine_sim_sum/(WT_G+WT_S+WT_T)
    you_watched=[]
    for j in np.where((cosine_sim_mean<1) & (cosine_sim_mean>0.42))[0].tolist():
        you_watched.append([docs_title[j],])

    if len(you_watched) > 0:
        c+=1
        chance_sim_all_watched = np.sum(cosine_sim_mean[np.where((cosine_sim_mean<1) & (cosine_sim_mean>0.42))])/len(cosine_sim_mean[np.where((cosine_sim_mean<1) & (cosine_sim_mean>0.42))])
        if chance_sim_all_watched < 0.6:
            MIGHT.append(f'\n{c}:\nYou MIGHT like "{animes_season.title[i]}"\n{chance_sim_all_watched*100:.2f}% chance of liking it due to others titles you watched:\n{you_watched[:5]}')
        if chance_sim_all_watched > 0.6:
            DEFINITELY.append(f'\n{c}:\nYou DEFINITELY gonna like "{animes_season.title[i]}"\n{chance_sim_all_watched*100:.2f}% chance of liking it due to others titles you watched:\n{you_watched[:5]}')
print("\nHere are the ones that you DEFINITELY gonna LIKE:\n ")
for D in DEFINITELY:
    print(D)
print("\nHere are the ones that you MIGHT LIKE:\n ")
for M in MIGHT:
    print(M)
```
